```python
from PIL import Image

NUM = 21
ims = []
for i in range(1,NUM+1):
    ims.append(Image.open("image_{:04d}.jpg".format(i)))
    
```

```python
ims[0].height
```

```python
height_th = 240
width_th = height_th * ims[0].width / ims[0].height
W = 5
H = 4

im_back = Image.new("RGB", (int(width_th * W), int(height_th * H)), color=(128, 128, 128))
```

```python
print(height_th, width_th)
```

```python
for i in range(H):
    for j in range(W):
        num = j + i * H
        print(num, i, j)
        im_th = ims[num].resize((int(width_th), height_th))

        im_back.paste(im_th, (int(j*width_th), i*height_th))
        
im_back
```

```python
im_back.save("thumbnail_elephant.jpg")
```

```python
FN = "thumbnail_elephant.jpg"
INTERVAL = 30 # sec

LENGTH_MOVIE = 60* 10 + 53 # sec
SPLIT_NUM = 20 # thumbnail picture num

print("WEBVTT\n")
for i in range(SPLIT_NUM):
    t_start = i / SPLIT_NUM * LENGTH_MOVIE
    minute_start = int(t_start//60)
    sec_start = t_start % 60

    t_end = (i+1) / SPLIT_NUM * LENGTH_MOVIE
    minute_end = int(t_end//60)
    sec_end = t_end % 60
    print("{:02d}:{:02.3f} --> {:02d}:{:.3f}".format(minute_start, sec_start, minute_end, sec_end))
    
    print("{}#xywh={},{},{},{}\n".format(
        FN, int((i%W) * width_th),
        int( (i//W) * height_th ), int(width_th), height_th
    ))
```
